```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import pymongo
%matplotlib inline
```

```python
# Connection MongoDB Database

myclient = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = myclient["employee-resign-prediction"]
mycol = mydb["train-data"]
```

```python
# Check 10 Data Cploumn and Value

df = pd.DataFrame(list(mycol.find()))
df.head(10)
```

```python
# Drop _id Column

df.drop("_id", axis='columns', inplace=True)
df.drop("EmployeeNumber", axis='columns', inplace=True)
df.dtypes
```

```python
# Checking Value Column Isnull

df.isnull().sum()
```

```python
df.Age = pd.to_numeric(df.Age)
df.DailyRate = pd.to_numeric(df.DailyRate)
df.DistanceFromHome = pd.to_numeric(df.DistanceFromHome)
df.Education = pd.to_numeric(df.Education)
df.EmployeeCount = pd.to_numeric(df.EmployeeCount)
df.EnvironmentSatisfaction = pd.to_numeric(df.EnvironmentSatisfaction)
df.HourlyRate = pd.to_numeric(df.HourlyRate)
df.JobInvolvement = pd.to_numeric(df.JobInvolvement)
df.JobLevel = pd.to_numeric(df.JobLevel)
df.JobSatisfaction = pd.to_numeric(df.JobSatisfaction)
df.MonthlyRate = pd.to_numeric(df.MonthlyRate)
df.MonthlyIncome = pd.to_numeric(df.MonthlyIncome)
df.NumCompaniesWorked = pd.to_numeric(df.NumCompaniesWorked)
df.PercentSalaryHike = pd.to_numeric(df.PercentSalaryHike)
df.PerformanceRating = pd.to_numeric(df.PerformanceRating)
df.RelationshipSatisfaction = pd.to_numeric(df.RelationshipSatisfaction)
df.StandardHours = pd.to_numeric(df.StandardHours)
df.StockOptionLevel = pd.to_numeric(df.StockOptionLevel)
df.TotalWorkingYears = pd.to_numeric(df.TotalWorkingYears)
df.TrainingTimesLastYear = pd.to_numeric(df.TrainingTimesLastYear)
df.WorkLifeBalance = pd.to_numeric(df.WorkLifeBalance)
df.YearsAtCompany = pd.to_numeric(df.YearsAtCompany)
df.YearsInCurrentRole = pd.to_numeric(df.YearsInCurrentRole)
df.YearsSinceLastPromotion = pd.to_numeric(df.YearsSinceLastPromotion)
df.YearsWithCurrManager = pd.to_numeric(df.YearsWithCurrManager)
```

```python
df.dtypes
```

```python
# Creating Array with Feature Column in Data to Convert

yes_no_columns = ['Attrition','OverTime']
```

```python
# Looping in Array to Replace Value Yes and No

for col in yes_no_columns:
    df[col].replace({'Yes': 1,'No': 0},inplace=True)
```

```python
df.dtypes
```

```python
df.head(5)
```

```python
df['Gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
for col in df:
    print(f'{col}: {df[col].unique()}')
```

```python
# Looping in Array to Replace Value Y

y_columns = ['Over18']
```

```python
# Looping to Reaplace Y -> 1

for col in y_columns:
    df[col].replace({'Y': 1},inplace=True)
```

```python
# Check Unique Column

for col in df:
    print(f'{col}: {df[col].unique()}')
```

```python
# Create Dummy Data from Value Column

df1 = pd.get_dummies(data=df, columns=['BusinessTravel','Department','EducationField','JobRole','MaritalStatus'])
df1
```

```python
# Check Data Type

df1.dtypes
```

```python
# Create Scala Array and Scala Function on Range Value

cols_to_scale = ['DailyRate','DistanceFromHome','EnvironmentSatisfaction','HourlyRate','JobInvolvement','JobLevel','JobSatisfaction','MonthlyIncome','MonthlyRate','NumCompaniesWorked','PercentSalaryHike','PerformanceRating','RelationshipSatisfaction','StandardHours','StockOptionLevel','TotalWorkingYears','TrainingTimesLastYear','WorkLifeBalance','YearsAtCompany','YearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df1[cols_to_scale] = scaler.fit_transform(df1[cols_to_scale])
```

```python
# Check Unique Column

for col in df1:
    print(f'{col}: {df1[col].unique()}')
```

```python
# Check Data Type

df1.dtypes
```

```python
X = df1.drop('Attrition',axis='columns')
y = df1['Attrition']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
import tensorflow as tf
from tensorflow import keras


model = keras.Sequential([
    keras.layers.Dense(52, input_shape=(52,), activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=500)
```
